```python
import numpy as np
import pandas as pd
import json
```

```python
df = pd.DataFrame(np.load('./data/npy/me_arr.npy')[250, :, :])
```

```python
df.to_json('my_csv', orient='records')
```

```python
df
```

```python
# // {xKey: 1, xLabel: "Astro.1", yKey: 1, yLabel: "Astro.1", val: 0.1904853455312878}
_cols = df.columns.values
_rows = df.index.values

dictlist = []
arr = df.values
for r in range(arr.shape[0]):
    for c in range(arr.shape[1]):
        data = {}
        data['xKey'] = c+1
        data['xLabel'] = _cols[c]
        data['yKey'] = r+1
        data['yLabel'] = _rows[r]
        data['val'] = df.iloc[r, c]
        dictlist.append(data)
```

```python
pd.DataFrame(dictlist).to_json('posterior_mu.json', orient='records')
```
